# Notebook Local - Carga de CSV.gz
Este notebook carga el archivo CSV.gz local y muestra las primeras 10 filas.


```python
import duckdb

# Crear conexión a DuckDB
con = duckdb.connect()

print("✅ Conexión a DuckDB establecida")
```

```python
# Cargar el CSV.gz en una tabla
con.execute("""
    CREATE OR REPLACE TABLE viajes_distritos AS 
    SELECT * FROM read_csv('20230301_Viajes_distritos.csv.gz', compression='gzip', all_varchar=true)
""")

print("✅ CSV.gz cargado exitosamente")
```

```python
# SELECT LIMIT 10
result = con.execute("SELECT * FROM viajes_distritos LIMIT 10").fetchdf()
result
```
